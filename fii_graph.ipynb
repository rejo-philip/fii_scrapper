```python

from plotly.subplots import make_subplots
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_scatter(secondary_y=False, x=fii_data['Date'], y=fii_data['Future Index Long'],
                marker_color="MediumSlateBlue", name="volume_data")

fig.add_scatter(secondary_y=True,  x=nifty_data['date'], y=nifty_data['close'],
                marker_color="DarkOrange", name="nifty_price_data")

fig.update_yaxes(secondary_y=False,
                 tickfont_color="MediumSlateBlue", title="Volume")

fig.update_yaxes(secondary_y=True,
                 tickfont_color="DarkOrange", title="index point")

fig.update_layout(title_text="A dual-Y-axis chart")
fig.show()


```
